### Remove outermost parantheses

A valid paratheses string is defined by the following rules  
- the string is empty
- If A is valid parantheses string (let A = "") then (A) is also valid like "()"
- If A and B are valid then A + B are also valid like ()()

Consider string "()()" -> We can split this into valid strings like "()" + "()" so string is non-primitive  

Consider string "(())" -> We cannot split it into valid paranthesis string like above so this is a primitive string.

Consider the following string "(()())(())". We have to remove the outer most paranthesis of the primitive strings like  
decompose -> "(()())" + "(())"  
- (()()) -> remove -> ()()
- (()) -> remove -> ()   

Combine afer removing -> ()()()

Consider the string s = "(()())(())"



output = "()()()"

```python
# Linear traversing
s = "(()())(())"
output = ""
c = 0
for i in range(len(s)):
    if s[i] == "(":
        c+=1
        if c>1:
            output+="("
    elif s[i] == ")":
        c-=1
        if c>0:
            output+=")"
print(output)
# Time complexity: O(n)
# Space complexity: O(1)
```

### Reverse words in a string
Consider the string s = 'the sky is blue'. This string contains words separated by ' ' (space) atleast one.

A word is defined as a sequence of non-space characters.

We have to return string where words are printed in reverse order  
like output = 'blue is sky the'

We have to remove trailing spaces and also extra space between the words.

lets create a list to store all the words extracted from the string. Words = []

To start extracting we have to loop through every character.

s = "hello world"

Start with index - 0

0 -> "h" non-space character -> word = "h"  
1 -> "e" non-space character -> word = "he"  
2 -> "l" non-space character -> word = "hel"  
3 -> "l" non-space character -> word = "hell"  
4 -> "o" non-space character -> word = "hello"  
5 -> " " space detected -> add word to Words -> reset word = "" (empty string)  
6 -> "w" non-space character -> word = "w"  
7 -> "o" non-space character -> word = "wo"  
8 -> "r" non-space character -> word = "wor"  
9 -> "l" non-space character -> word = "worl"  
10 -> "d" non-space character -> word = "world"

End of string. Add word to list after loop ends.

Now reverse the list.

And add each word to string output using loop.

```python
# Linear approach
s = "hello world"
words = []
output = ""
word = ""
for i in s:
    if i != " ":
        word+=i
    elif word:
        words.append(word)
        word = ""

if word:
    words.append(word)
words.reverse()

for i in range(len(words)):
    output += words[i]
    if i < len(words)-1:
        output+=" " # add space at end if the word is not the last

print(output)
# Time complexity: O(n)
# Space complexity: O(n) worst case senario
```

We can avoid the list and start the string from the last from back side. We can set a point to track the spaces.

s = "hello world"

set pointer i at last index, if i is space or a not a character keep moving back with inner loop until a character is encounter.

Keep a safety check if i < 0

When i is a character here i = "d" set another pointer j at that index, keep moving i back untill a space is found.

Set word = s[i+1:j+1]

if output is not empty and extra space first and then ad the word to the output string.

```python
s = "hello world"
words = []
output = ""
word = ""
i = len(s)-1
j = i
while i>=0:
    # Skipping spaces
    while i >=0 and s[i] == ' ':
        i-=1

    # break loop if i < 0
    if i < 0:
        break

    # set another pointer
    j = i

    # skip characters until space is found
    while i>=0 and s[i]!=' ':
        i-=1

    word = s[i+1:j+1]

    # if output string is not empty add a space
    if output != "":
        output += " "

    # add word to the string
    output += word

print(output)
# Time complexity: O(n)
# Space complexity: O(1)
```
