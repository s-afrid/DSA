## Quick Sort
Consider the unsorted array: [4,6,2,5,7,9,1,3]  
The first step of quick is -> Pick a pivot and place it in its correct place in the sorted array  
The pivot element can be 1st element, last element, median element or any random element of the array.  

In [4,6,2,5,7,9,1,3] -> 4 is the pivot

### Code logic
This logic explains how to put the pivot at the right place and making lower elements to the right and higher elements to the left.  


The pivot now has become a partition index dividing higher and lower elements.  
![Screenshot from 2025-08-10 22-25-09.png](<attachment:Screenshot from 2025-08-10 22-25-09.png>)  
Repeat the same for lower:partition-1 and partition+1:high

```python
def get_pivot(arr,low,high):
    pivot = arr[low]
    i,j = low+1,high
        
    while 1:
        # move i till i is greater than pivot
        while arr[i] < pivot and i<=high:
            i+=1
        # move j till j is less than pivot
        while arr[j] > pivot and j>=low:
            j-=1

        # stop when j crosses i
        if j<i:
            break
        
        # swap when i and j both stops
        arr[i],arr[j] = arr[j],arr[i]

    # swap with j when j crosses i
    arr[j],arr[low] = arr[low],arr[j]
    return arr.index(pivot)
    

def quick_sort(arr,low,high):
    if high>low:
        # get the index of pivot after placing it at the correct place
        pindex = get_pivot(arr,low,high)
        # repeat for left side
        quick_sort(arr,low,pindex-1)
        # repeat for right side
        quick_sort(arr,pindex+1,high)
        
arr = [4,6,2,5,7,9,1,3]
print(arr)
quick_sort(arr,0,len(arr)-1)
print(arr)



```
